# Preparing Data for model

```python
# Get names of columns with missing values
cols_with_missing = [col for col in data_train.columns
                     if data_train[col].isnull().any()]
print("Columns with missing values:")
print(cols_with_missing)
```

```python
#we are going to enter the mean values in respective null columns
data_train['Seats'].fillna(data_train['Seats'].mean(),inplace=True)
data_test['Seats'].fillna(data_test['Seats'].mean(),inplace=True)
```

To achieve more accurate results we need more data to work with.
Therefore we are going to merge the two sets of data given to us

```python
data = pd.concat([data_train,data_test], sort=False)
```

```python

plt.figure(figsize=(20,5))
data_train['Mileage'].value_counts().head(100).plot.bar()
plt.show()
```

NOTE: In case of mileage we are going to fill NaN values with mode i.e 17 kmpl
Also I observed that some of the cars have 0.0kmpl mileage which is not practical therefore I am changing 0.0 kmpl to 17 kmpl

```python
data_train['Mileage'] = data_train['Mileage'].fillna('17.0 kmpl')
data_test['Mileage'] = data_test['Mileage'].fillna('17.0 kmpl')


data_train['Mileage'] = data_train['Mileage'].replace("0.0 kmpl", "17.0 kmpl")
data_test['Mileage'] = data_test['Mileage'].replace("0.0 kmpl", "17.0 kmpl")
```

```python
plt.figure(figsize=(20,5))
data['Engine'].value_counts().head(100).plot.bar()
plt.show()
```

It is clear that for engines mode value is 1197cc. Therefore we are going to fill 1197cc for NaN values in Engine column

```python
data_train['Engine'] = data_train['Engine'].fillna('1197 CC')
data_test['Engine'] = data_test['Engine'].fillna('1197 CC')
```

```python
plt.figure(figsize=(20,5))
data['Power'].value_counts().head(100).plot.bar()
plt.show()
```

we can see that mode for Power column is 74bhp.
Therefore we can fill 74 bhp in place for Nan value

Also I observed that some of the entries in power column contains "null bhp" as entries
Therfore we need to replace that as well with 74 bhp

```python
data_train['Power'] = data_train['Power'].fillna('74 bhp')
data_test['Power'] = data_test['Power'].fillna('74 bhp')


data_train['Power'] = data_train['Power'].replace("null bhp", "74 bhp")
data_test['Power'] = data_test['Power'].replace("null bhp", "74 bhp")
```

```python
data_train.isnull().sum()
```

```python
data.info()
```

To be able to utilise this data we need to convert power, engine , mileage to float ,float ,float respectively

```python
data_train['Mileage'] = data_train['Mileage'].apply(lambda x : str(x).split(' ')[0]).astype(float)
data_test['Mileage'] = data_test['Mileage'].apply(lambda x : str(x).split(' ')[0]).astype(float)
```

```python
data_train['Engine'] = data_train['Engine'].apply(lambda x : str(x).split(" ")[0]).astype(float)
data_test['Engine'] = data_test['Engine'].apply(lambda x : str(x).split(" ")[0]).astype(float)
```

```python
data_train['Power'] = data_train['Power'].apply(lambda x : str(x).split(' ')[0]).astype(float)
data_test['Power'] = data_test['Power'].apply(lambda x : str(x).split(' ')[0]).astype(float)
```

```python
data_train['Year'] = data_train['Year'].astype(int)
data_train['Location'] = data_train['Location'].astype('category')
data_train['Fuel_Type'] = data_train['Fuel_Type'].astype('category')
data_train['Transmission'] = data_train['Transmission'].astype('category')
data_train['Owner_Type'] = data_train['Owner_Type'].astype('category')

data_test['Year'] = data_test['Year'].astype(int)
data_test['Location'] = data_test['Location'].astype('category')
data_test['Fuel_Type'] = data_test['Fuel_Type'].astype('category')
data_test['Transmission'] = data_test['Transmission'].astype('category')
data_test['Owner_Type'] = data_test['Owner_Type'].astype('category')
```

```python
data_train.info()
```

```python
from sklearn.model_selection import train_test_split

y = np.log1p(data_train.Price)  
X = data_train.drop(['Price'],axis=1)


X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size=0.82,test_size=0.18,random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()


X_train['Name'] = label_encoder.fit_transform(X_train['Name'])
X_valid['Name'] = label_encoder.fit_transform(X_valid['Name'])
data_test['Name'] = label_encoder.fit_transform(data_test['Name'])

X_train['Location'] = label_encoder.fit_transform(X_train['Location'])
X_valid['Location'] = label_encoder.fit_transform(X_valid['Location'])
data_test['Location'] = label_encoder.fit_transform(data_test['Location'])

X_train['Fuel_Type'] = label_encoder.fit_transform(X_train['Fuel_Type'])
X_valid['Fuel_Type'] = label_encoder.fit_transform(X_valid['Fuel_Type'])
data_test['Fuel_Type'] = label_encoder.fit_transform(data_test['Fuel_Type'])

X_train['Transmission'] = label_encoder.fit_transform(X_train['Transmission'])
X_valid['Transmission'] = label_encoder.fit_transform(X_valid['Transmission'])
data_test['Transmission'] = label_encoder.fit_transform(data_test['Transmission'])

X_train['Owner_Type'] = label_encoder.fit_transform(X_train['Owner_Type'])
X_valid['Owner_Type'] = label_encoder.fit_transform(X_valid['Owner_Type'])
data_test['Owner_Type'] = label_encoder.fit_transform(data_test['Owner_Type'])
```

```python
X_train.head()
```

# MODEL

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_squared_log_error

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)

predictions = my_model.predict(X_valid)
print("MAE: " + str(mean_absolute_error(predictions, y_valid)))
print("MSE: " + str(mean_squared_error(predictions, y_valid)))
print("MSLE: " + str(mean_squared_log_error(predictions, y_valid)))
```

# PREDICTION

```python
preds_test = my_model.predict(data_test)
preds_test = np.exp(preds_test)-1 #converting target to original state
print(preds_test)

preds_test = preds_test.round(2)
print(preds_test)
```

```python
output = pd.DataFrame({'Price': preds_test})
output.to_excel('submission.xlsx', index=False)
```
